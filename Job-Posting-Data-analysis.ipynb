# Data Visualization of Number of applicant vs Job title

```python
# Group by Job title
df['applies'] = pd.to_numeric(df['applies'], errors='coerce')

job_applicants = df.groupby('title')['applies'].sum().reset_index()

job_applicants = job_applicants.sort_values(by='applies', ascending=False)

print(job_applicants)
```

```python
# Visualize applicants per job title
import matplotlib.pyplot as plt

# Plot top 10 job titles with the highest applicants
top_jobs = job_applicants.head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_jobs['title'], top_jobs['applies'], color='skyblue')
plt.xlabel('Number of Applicants')
plt.ylabel('Job Title')
plt.title('Number of Applicants for Top 10 Job Titles')
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```

# Data Visualization of most Job offered by Company

```python
# Remove rows with null values in 'company_id'
df_cleaned = df.dropna(subset=['company_id'])

# Replace null values in 'company_id' with 'Unknown'
df['company_id'] = df['company_id'].fillna('Unknown')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by 'company_id' and count the number of jobs per company
company_job_count = df.groupby('company_id').size().reset_index(name='job_count')

# Sort by the number of jobs and take the top 10
top_10_companies = company_job_count.sort_values(by='job_count', ascending=False).head(10)

# Plot histogram of job counts for top 10 companies
plt.figure(figsize=(10, 6))
sns.histplot(top_10_companies['job_count'], bins=10, kde=False, color='skyblue')

# Add labels and title
plt.title('Histogram of Job Counts for Top 10 Companies', fontsize=16)
plt.xlabel('Number of Jobs per Company', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.show()
```

# Visualization of applicant vs Salary

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'applies' and 'max_salary' columns to numeric values
df['applies'] = pd.to_numeric(df['applies'], errors='coerce')
df['max_salary'] = pd.to_numeric(df['max_salary'], errors='coerce')

# Drop rows where 'applies' or 'max_salary' are null
df_clean = df.dropna(subset=['applies', 'max_salary'])

# Create a scatter plot of applies vs max salary
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_clean, x='max_salary', y='applies', color='skyblue')

# Add labels and title
plt.title('Applicants vs Maximum Salary', fontsize=16)
plt.xlabel('Maximum Salary', fontsize=12)
plt.ylabel('Number of Applicants', fontsize=12)

# Show the plot
plt.show()
```
